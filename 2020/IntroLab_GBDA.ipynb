***B1) Lists***



````
# Create a list
a = [1, 2, 3, 4]
print(a[0])
# Loops
animals = {'cat', 'dog', 'fish'}
for idx, animal in enumerate(animals):
    print('#%d: %s' % (idx + 1, animal))

# Lists can contain different types of data as elements 
a[1] = 'bill'
print(a)

# Print last element of list a
print(a[-1])

# Append a new element at the end of the list
a.append('70')
print(a)

# Remove and return the last element of the list
ab = a.pop()
print(ab)
````

````
# List slicing 
a = list(range(8))      # create a list of integers
print(a)
print(a[1:3])           # last element is excluded!
print(a[:])             # : suggests to all elements
print(a[:-1])           # print all elements except the last one
a[3:4] = [5, 6]         # change a slice of a list
print(a)
````

````
# Loops
colors = ['orange', 'yellow', 'blue', 'red']
for color in colors:
  print(color)

# How can I get the index of each element within a loop ?
for idx, color in enumerate(colors):
  print('The color {} has the index {}'.format(color, idx))

# List comprehensions
numbers = list(range(10))
print(numbers)
even_numbers = [x for x in numbers if x % 2 == 0]
print(even_numbers)
````

***B2) Dictionairies***

````
# Dictionairies store pairs of (idx/key, value)

# Unlike lists, a new dictionairy key can also be assigned a value, 
# not just ones that already exist!

d = {0:'athena', 1:'bill', 2:'kostas'}
print(d[0])

d = {'athena':'blonde', 'bill':'happy'}
print(d['bill'])

# set an element on a dictionairy
d['cat'] = 'pablo'

# check if a dictionairy has a given key (boolean)
print('athena' in d)
print('margarita' not in d)

# get method does the same thing as indexing....
print(d.get('athena'))
print(d.get(8))  # but if false returns None 
````

***B3) Sets***

````
# Sets is a collection which is unordered and unindexed
myset = {'red', 'green', 'blue'}
print(myset)

for color in myset:
  print(color)

# Add vs Update methods
myset.add('yellow')
print(myset)

myset.update(['magenta', 'red', 'cyan'])
print(myset)

# Union and intersection methods
myset2 = {'banana', 'apple'}
myset3 = {'orange', 'apple'}

print(myset2.union(myset3))
print(myset2.intersection(myset3))
````

***B4) Tuples***


````
# Tuples are immutable ordered list of values (*cannot be changed!)
d = (1, 2, 3)
print(d)

dictionary = {d: x for x in range(5)}
print(dictionary[d])

# Trying to reassign a value in a tuple causes TypeError
try:
  d[0] = 'athena'
except TypeError:
  print('Changing a tuple causes a TypeError')
````

## **C)   Functions and Classes**

````
# We define functions using the def keyword 
def is_even(x):
  if x % 2 == 0:
    return('The number is even')
    # Everything after the return statement will never happen!
    print('This will not be printed')
  else:
    return('The number is odd')

num = list(range(10))
for idx in num:
  print(is_even(idx))
# You must define functions before they are called, 
# in the same way that you must assign variables before using them.


# You can also define functions with more than one argument
# Separate them with commas
def print_sum_twice(x1, y1):
   print(x1 + y1)
   print(x1 + y1)

print_sum_twice(5, 8)


# Warning : 
# Function arguments cannot be referenced outside of the function's definition!
print(x1)
````

````
# This is how we define a class in python 

class Greeter(object):

    # Constructor
    def __init__(self, name):
        self.name = name  # Create an instance variable

    # Instance method
    def greet(self, loud=False):
        if loud:
            print('HELLO, %s!' % self.name.upper())
        else:
            print('Hello, %s' % self.name)

g = Greeter('Fred')  # Construct an instance of the Greeter class
g.greet()            # Call an instance method; prints "Hello, Fred"
g.greet(loud=True)   # Call an instance method; prints "HELLO, FRED!"
````

### **Array manipulation**

````
Q = np.random.random((2, 4))

print('Q\n', Q)
print('Size of Q:\n', Q.shape)

# Slicing
print('Slice 1\n', Q[:, 1:2])
print('Slice 2\n', Q[1:, :])

# Indexing
print('Element (0,2) of Q:\n', Q[0, 2])

# Permutation
print('Q permutated:\n', Q[(1, 0)])

# Reshape
print('Q as (4,2):\n', Q.reshape([4, 2]))

print('"Flattened" Q:\n', Q.flatten())
````

### **Linear Algebra**

````
a = np.random.randint(5, size=(2,4))
b = np.random.randint(5, size=(2,4))
print('a\n', a, '\nb\n', b)

# Sum
print(np.sum(a))
print(np.sum(a, axis=1))
# Inner product
c = np.dot(a, b.T)
print(c)

# Inversion of a matrix
d = np.linalg.inv(c)
print(d)

````

````
import torch
import numpy as np
````

````
# What is a tensor?
# ... basically generalization of the concept of matrix to N>=3 dimensions
np_tensor = np.array([[[1, 2, 3],
               [4, 5, 6]],
              [[10, 20, 30],
               [40, 50, 60]]])
print(np_tensor)
print('This tensor is of rank %d' %(np_tensor.ndim))
````

````
# Create a random Pytorch tensor 
tensor = torch.rand(5,3).to(torch.float32)
print('Random tensor\n', tensor, f'\n with size: {tensor.size()}')
````

````
# Create a random-tensor based on an existing tensor
x = torch.randn_like(tensor, dtype=torch.float)
print(x, '\n', x.size())
# Result has the same size as before...
````

````
# Methods will reuse properties of the input tensor unless new values are provided by the user
x = tensor.new_ones(5, 3, dtype=torch.float)
print(x)
````

````
print(x.size())
````

**Useful tensor operations**

````
t = torch.rand(2,2)
t
````

````
t.max()
````

````
# Use .item() to get the value as a Python number
t.max().item()
````

````
# Addition operation
tt = torch.rand(2,2)
print('tt\n', tt)
print('t+tt\n', t+tt)

print('torch.add(t, tt)  -- SAME\n', torch.add(t, tt))
````

````
### Note : Every operation that ends with _ will mutate the tensor in-place
# Expample :Add tt to t
tt.add_(t)
print('tt.add_(t) --SAME but in-place\n', tt)
````

````
# We can use standard NumPy-like indexing 
print(tt[:,0])
````

````
# Tensor resizing 
x = torch.arange(36)
print('x is a_range vector! \n', x)
y = x.view(6, 6)
print('y is x data in a 6x6 matrix! \n', y)
z = x.view(-1, 12)
print('z is x data in a 3x12 matrix!\nThe number of rows (3) is automatically computed to fit the data! \n', y)
print('Sizes: {x, y, z}\n', x.size(), y.size(), z.size())
````

**Convert NumPy array to Torch tensor**

````
import numpy as np
a = np.ones(5)
b = torch.from_numpy(a)
np.add(a, 1, out=a)
print('a is a numpy array!', a)
print('b is a Pytorch Tensor now!', b)
````

````
# Let's do this in other ways!

# Create a Pytorch Tensor from numpy array (by ref)
c = torch.as_tensor(a)
print('b -- SAME as BEFORE!\n', c)

# Copy-create a Pytorch tensor from numpy array
d = torch.tensor(a)
print('c\n', d)

print('Changing np array a (in-place by adding 10!)......')
np.add(a, 10, out=a)
print('New numpy array a\n', a)
print('Tensor by ref\n', c)
print('Tensor by copy\n', d)
print('Last, ... Tensor "b" from BEFORE\n', b)
````
